# Model overfitting
- Scenario with large training accuracy but less testing accuracy

```python
# Load relevant scientific libraries.
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline
np.random.seed(42)

# Create m=100 samples for Polynomial Regression

m = 100

X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)

# Plot the data for visualization.
plt.plot(X, y, 'bo')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Samples for Polynomial Regression')

plt.show()
```

## Higher degree polynomial fit
- degree = 30

```python
# With Polynomial Regression
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=30, include_bias=False)

X_poly = poly_features.fit_transform(X)

# Fit linear regression on polynomial features

poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)


# A set of samples for looking at visualization of Linear Regression

X_new = np.linspace(-2.5, 2.5, 100).reshape(100,1)

# Predict using polynomial features
X_new_poly = poly_features.transform(X_new)
y_pred_poly= poly_reg.predict(X_new_poly)

# Plot the data for visualization.
plt.plot(X, y, 'bo')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Samples for Polynomial Regression')

plt.plot(X_new, y_pred_poly, 'r-', linewidth=2, label='Polynomial Regression')

plt.show()
```

## Regularization strategies to overcome overfitting scenario

- $L1$ (LASSO: Least Absolute Shrinakge and Selection Operator)

  - $J(\theta) = MSE(\theta) + 2\alpha \sum_{i=1}^{n} abs(\theta_i) $
  - Eliminates least important features by setting weight to zero.

- $L2$ (Ridge)

  - $J(\theta) = MSE(\theta) + \frac{\alpha}{m} \sum_{i=1}^{n} \theta_i^2 $
  - Keeps the model weights as small as possible

- Elastic Net

  - $J(\theta) = MSE(\theta) + r \left(2\alpha \sum_{i=1}^{n} abs(\theta_i) \right)+ (1-r) \left(\frac{\alpha}{m} \sum_{i=1}^{n} \theta_i^2 \right)$


```python
# L1-Regularization
from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X, y)
```

```python
# L2-Regularization
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha=1, solver="cholesky")
ridge_reg.fit(X, y)
```

```python
from sklearn.linear_model import ElasticNet
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic_net.fit(X, y)
```
